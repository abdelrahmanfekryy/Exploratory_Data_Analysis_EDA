# Match_Player

```python
Match_Player = Match[['match_api_id']].copy()

for status in ["away","home"]:
    for i in range(1,12):
        Match_Player[f"{status}_{i}"] = Match[f'{status}_player_{i}'].astype(str) + "_" + Match[f'{status}_player_X{i}'].astype(str) + "_" + Match[f'{status}_player_Y{i}'].astype(str)

Match_Player = pd.melt(Match_Player, id_vars=['match_api_id'], var_name='attrib', value_name='value')
Match_Player[['player_api_id','x_coord','y_coord']] = Match_Player['value'].str.split('_',expand=True)
Match_Player[['team_status','N']] = Match_Player['attrib'].str.split('_',expand=True)
Match_Player.drop(labels=['N','value','attrib'],axis=1,inplace=True)   
```

```python
Match_Player
```

# Match_Team

```python
Match_Team = Match[['match_api_id']].copy()

for status in ["away","home"]:
    Match_Team[f"{status}"] = Match[f'{status}_team_api_id'].astype(str) + "_" + Match[f'{status}_team_goal'].astype(str)

Match_Team = pd.melt(Match_Team, id_vars=['match_api_id'], var_name='team_status', value_name='value')

Match_Team[['team_api_id','goal']] = Match_Team['value'].str.split('_',expand=True)
Match_Team.drop(labels=['value'],axis=1,inplace=True)  
Match_Team 
```

# League/Country

```python
LeagueData = League.set_index('country_id').join(Country.set_index('id'), how='left',lsuffix='_league', rsuffix='_country').reset_index()
LeagueData.drop(labels=['country_id'],axis=1,inplace=True)
LeagueData.rename(columns={'id':'league_id','name_league':'league_name','name_country':'country_name'},inplace=True)
LeagueData
```

```python
with pd.ExcelWriter('Data/European_Soccer_Database.xlsx', engine='xlsxwriter') as writer:
    Player_Attributes.to_excel(writer,sheet_name='Player_Attributes',index=False)
    Player.to_excel(writer,sheet_name='Player',index=False)
    Match.to_excel(writer,sheet_name='Match',index=False)
    Match_Player.to_excel(writer,sheet_name='Match_Player',index=False)
    Match_Team.to_excel(writer,sheet_name='Match_Team',index=False)
    LeagueData.to_excel(writer,sheet_name='League',index=False)
    Team.to_excel(writer,sheet_name='Team',index=False)
    Team_Attributes.to_excel(writer,sheet_name='Team_Attributes',index=False)
```
