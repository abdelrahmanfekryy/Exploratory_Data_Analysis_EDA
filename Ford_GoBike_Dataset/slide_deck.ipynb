# Ford GoBike Dataset Insights.

In This Presentation We Are Going To Cover Two Main Points:

* The Detailed Relationship Between Member Gender And Trip Duration.
* The Detailed Relationship Between Member User Type And Trip Duration.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from plot_assist import *
df = pd.read_csv('data/fordgobike_tripdata_clean.csv')
```

## Who Takes Longer Rides Customers or Subscribers?

```python
box_plot(df,'user_type','duration_sec',ylim=(0,2500),showfliers=False)
```

## we can see that customers have longer rides on average.

## the next plot convey the same insight from different perspective

```python
plt_hist(df['user_type'],df['duration_sec'],bins=2000,scale='linear',xlim=(61,4000),histtype='step')
```

## Now That We Know That Customers Have Longer Rides ...
## Is That Longer Rides Are Distributed Equally Through All Days ? 

```python
order = ['Saturday', 'Sunday','Monday','Tuesday','Wednesday','Thursday','Friday']
box_plot_multi(df,'start_weekday','duration_sec','user_type',ylim=(0,3500),showfliers=False,order=order)
```

## we can see that Customers longer rides are more concentrated on weekend

## Is That Longer Rides Are Distributed Equally Through All Day Hours ?

```python
box_plot_multi(df,'start_hour','duration_sec','user_type',ylim=(0,3500),showfliers=False)
```

## we can see that Customers longer rides are concentrated during business hours and at the early hours of the day

## Who Takes Longer Rides Males, Females Or Other Genders ?

```python
box_plot(df,'member_gender','duration_sec',ylim=(0,2000),showfliers=False)
```
